#### This file generates two files:
* action_frequency: each row in the table represents a time a user like/dislike/wishlist.
* action_watchmorethanhalf: each row in the table represents a time when a user watches more than half of a trailer

```python
import pandas as pd
import numpy as np
```

```python
action_data_raw = pd.read_csv('../Clean_Data/log_trailer_actions_sessionIds.csv')
```

```python
# Get the algorithm assignment
expt_trailer = pd.read_csv('../Data/expt_trailer.csv', sep=',')
alg_assignment = expt_trailer[['userId','loginId','recommenderId']]
alg_assignment.sort_values(by='recommenderId').shape
```

```python
action_alg_data_raw = pd.merge(action_data_raw, alg_assignment, on=['loginId','userId'], how='left')
```

```python
action_alg_data_raw.dropna(inplace=True)
```

```python
action_alg_data_raw.groupby("action").count()
```

```python
action_alg_data = action_alg_data_raw[action_alg_data_raw['action'].isin(["TrailerLiked", "TrailerDisliked", "WishlistedMovie","MovieDetailsModalOpened","GoToMovieDetails"])]
```

```python
action_alg_data.head()
```

#### Include actions taken on seed movies by appending information about them to the table: ⬇️

```python
all_recs_info = pd.read_csv('../Data/rec_seed_info_raw.csv')
```

```python
row_0 = all_recs_info.loc[0].copy()
```

```python
new_df_list = []
new_df_list.append(row_0)

for i in range (0, len(all_recs_info)):
    if (all_recs_info.loc[i]["SeedMovie"]) != new_df_list[len(new_df_list)-1]["SeedMovie"]:
        temp_row = all_recs_info.loc[i].copy()
        temp_row["movieId"]=temp_row["SeedMovie"]
        new_df_list.append(temp_row)
    new_df_list.append(all_recs_info.loc[i])
```

```python
all_recs_include_seed_info = pd.DataFrame(new_df_list)               
```

```python
all_recs_include_seed_info
```

#### Complete appending information about seed movies to the table ⤴️

#### Load RC trailers & alg assignment

```python
rec_played_alg = pd.read_csv("../Clean_Data/rec_played_alg.csv")
rec_played_alg.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
rec_played_alg.dropna(inplace=True)
```

```python
rec_content_include_seed = pd.merge(rec_played_alg,all_recs_include_seed_info,on=["loginId","sessionIds","movieId"],how="left")
```

```python
rec_content_include_seed.dropna(inplace=True)
```

```python
rec_interactions = pd.merge(rec_content_include_seed,action_alg_data,on=["loginId","sessionIds","movieId"],how="left")
rec_interactions.drop(["position_y","recommenderId_y"], axis=1, inplace=True)
```

```python
rec_interactions.count()
```

```python
rec_interactions['isLiked'] = False
rec_interactions['isDisliked'] = False
rec_interactions['isWishlisted'] = False
```

```python
for index, row in rec_interactions.iterrows():
    if row['action_y']=="TrailerLiked":
        #row.['isLiked']=True
        #row.isLiked=True

        rec_interactions.loc[index,'isLiked']=True
    elif row['action_y']=="TrailerDisliked":
        #row['isDisliked']==True
        rec_interactions.loc[index,'isDisliked']=True
        #row.isLiked=True
    elif row['action_y']=="WishlistedMovie":
        if (row['tstamp_y']>row['tstamp_x']):
            #row['isWishlisted']==True
            rec_interactions.loc[index,'isWishlisted']=True
            #row.isLiked=True
    
```

```python
rec_interactions.loc[rec_interactions.action_y=="TrailerLiked",'isLiked']=True
rec_interactions.loc[rec_interactions.action_y=="TrailerDisliked",'isDisliked']=True
rec_interactions.loc[rec_interactions.action_y=="WishlistedMovie",'isWishlisted']=True
```

```python
rec_interactions.drop(['Unnamed: 0','userId_y','Algorithm','age','age_seedmovie'],axis=1,inplace=True)
```

```python
rec_interactions.to_csv('../Clean_Data/action_frequency_5_02.csv')
```
